- What is statistics ?
    - Collection, organizing, analysis, interpretation and presentation of data to enable better decision making.
    - 
- What are the types of statistics ?
    - Descriptive
        - Here the whole population is considered, we are organizing and summarizing the existing content.
            - Measure of central tendancy(how far the data is aligned with central tendancies)
                - Mean
                - median
                - mode
            - Measure of dispersion (how far the data points are diverged)
                - variance
                - standard deviation
    - Inferential 
        - Here a sample of a population is taken and made some inferences with some experiments(z-test, t-test) created, these inferences and conclusions made are implemented on other datasets(population data).

- what is population and what is sample ?
    - Population (Notation : N) : Includes all the specimens within the selected group of study.
    - Sample (Notation : n): A portion of the sample extracted to make some insights or inferences then validating the same with population.
    - Example to understand : Lets assume a media channel in India is trying to predict the outcome of an election. It is not possible for us collect the opinions of 120 billion and consolidate to answer. Thus they select a group which tries to replicate the diversity of the population.


- What is the measure of central tendancies ?
    - Here we are trying to understand the behaviour of majority of the population.
    - The metrics includes mean, median, mode.
        - Mean : average of the population or sample.
            - Since it calculates and includes all elements they are vulnerable to outliers.
        - median : the 50% th value of the distribution.
            - This requires sorting of elements in ascending or descending order
            - Then pick up the middle position of the orders.
            - This gives insights about how the 50% of the value behaves.
            - This is not completely vulnerable to outlier, but effectively handles to a point.
        - Mode : most occurring value within the population or sample

- What is the measure of dispersion ?
    - Variance (notation : sigma square)
        - We are trying to find the distance between the mean and average distance between all the other elements.
        - variance gives the spread of the data, however outlier highly impacts it.
        - This explains how far it is from central value, very high distance leads to outlier or anamoly, and too much closeness to central tendancies leads to pattern of general population behaviour.

- What is a variable ?
    - Variable is an instance that can take any value in any data format.
    - Most of the variables has two components : a verb --> methods and a noun --> parameters or features.
    - Variable is classified into 2 :
        - Qualitative(categorical data) like colors, high,medium and low, 
        - Quantitative
            - Discrete : These are generally numbers, counts, and definite numbers and only few values or easily countable or this is an whole number or has finite whole number values
                - Eg: Number of students in class
            - Continuous : has decimals values and non easily countable, these are ranges
                - Eg. Range of weights like 15.5kg . 16.5 kg



- what do you mean random variable ? (notation : X)
    - This is functions which has range of values from an experiment or an event. Eg. Tossing a coin, rolling a dice.
    - Types of random variable
        - Discrete random variable : eg --> tossing a coin or rolling a dice
        - continuous random variable : eg --> predicting the height of person attending a function tomorrow.

- What is percentile and quartiles ?
    - Percentile : is a value that represents the number of observations that lie below the specified percentage --> 
        - percentile of value x = (number of values below x / Count of sample size)*100
        - This depicts the % of values lie below the x value.
    
    - Quartile : values percent on the 25% or 50% or any other.
        - This states how the values lie within each quartile 
            - 25% - 1st Quartile
            - 50% - 2nd Quartile
            - 75% - 3rd Quartile

        - IQR = Q3 vale - Q1 value
        - Q1 = (25/100)*(n+1) where n is sample size 
        - Q3 = (75/100)*(n+1) where n is sample size 
        - lower fence = Q1 - 1.5*(IQR)
        - higher fence = Q3 + 1.5*(IQR)

        - Any value we considering outlier is us who are considering they are outlier.
        - If possible we have to understand the cause for the outlier.
        - Box plot enables the user to visualize the outliers and Whisker box plot also helps us regarding the same.


- Definition of co-variance and correlation ? 
    - These are measures used to determine the strength of relationships among two different random variable. Also, defines how one variable is related to other.
        - Co-variance :
            - This defines how two variables interact with each other. If positive co-variance then both of the variable tends to increase together else one variable increase while the other decreases.
            
           -  
           - if x increase y increase or x decrease y decrease = (+) co-variance
    - *Advantage of using co-variance*
        - Quantifies the relationship strength
    - *Disadvantage of using co-variance*
        - There is no specific limit value
        - Thus to pull these numbers within limits we will be using correlation
