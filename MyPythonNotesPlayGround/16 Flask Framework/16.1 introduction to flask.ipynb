## Flask Introduction

- What is flask and how is it used ? 
    - A framework used for creating web applications end to end.
    - This framework is built using the python programming language


    - How framework is different from library ?
        - Lets image the objective is to build a house. We need few inputs to complete the objective such as materials, tools, plan, standards, and instruction etc. In case of library, it is just a tool box that lets you to complete a small portion of work easily. This does not have any instruction prior on house to use them for building houses. So, the tools can be used for household applications, gardening application etc. Since they do not contain any specific instruction, a house construction could be hindered with many errors, iterations and time waste. However on flip side, let assume we are provided with a pre-fabricated kits, tools, instructions, guidelines,plan and execution manual. In this case error rate is low, low repetition and re-iterations needed. Since the details are as nuanced as color of walls, how should it be placed, sequence of construction etc.
        - In library, you are responsible for how are you going to use it. So, the objective completion effectiveness depends on your capability.
        - In framework, since the houses are built multiple times using the guidelines(rules and instruction) which are built by iteration(errors + corrections). The effectiveness of completion is high and error rate is low.
        - Module: A single tool (like a hammer) that provides specific functionality. It can contain multiple functions, classes, and variables.
        - Class: A blueprint for creating objects (like a design for a hammer). It defines how the objects should behave and what properties they have.
